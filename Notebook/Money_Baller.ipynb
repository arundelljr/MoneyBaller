```python
import pandas as pd
import numpy as np


# Load the CSV
df = pd.read_csv("../raw_data/FC26_20250921.csv", low_memory=False)


df.head()
```

```python
# Columns to drop
drop_cols = [
    # Metadata / URLs / club info
    'player_url','player_face_url',
    'fifa_version','fifa_update','fifa_update_date',
    'league_id','league_name','league_level','club_team_id','club_name',
    'club_position','club_jersey_number','club_loaned_from','club_joined_date',
    'club_contract_valid_until_year','nationality_id','nation_team_id',
    'nation_position','nation_jersey_number','player_tags','player_traits',
    'body_type','real_face',

    # Position-specific ratings
    'ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw',
    'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm',
    'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb', 'gk'
]

# Drop columns safely
df = df.drop(columns=drop_cols, errors='ignore')

# Quick check of remaining columns
df.columns
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
# Drop work_rate (all missing)
df = df.drop(columns=['work_rate'])

# Fill numeric skill and value columns with median
fill_median_cols = [
    'release_clause_eur', 'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic'
]
for col in fill_median_cols:
    df[col] = df[col].fillna(df[col].median())

# For goalkeeping_speed, set missing values to 0 (most players are not goalkeepers)
df['goalkeeping_speed'] = df['goalkeeping_speed'].fillna(0)

# Verify missing values
df.isnull().sum()
```

```python
import pandas as pd

# List of skill columns
skill_cols = [
    'overall','potential','pace','shooting','passing','dribbling','defending','physic',
    'attacking_crossing','attacking_finishing','attacking_heading_accuracy','attacking_short_passing','attacking_volleys',
    'skill_dribbling','skill_curve','skill_fk_accuracy','skill_long_passing','skill_ball_control',
    'movement_acceleration','movement_sprint_speed','movement_agility','movement_reactions','movement_balance',
    'power_shot_power','power_jumping','power_stamina','power_strength','power_long_shots',
    'mentality_aggression','mentality_interceptions','mentality_positioning','mentality_vision','mentality_penalties','mentality_composure',
    'defending_marking_awareness','defending_standing_tackle','defending_sliding_tackle',
    'goalkeeping_diving','goalkeeping_handling','goalkeeping_kicking','goalkeeping_positioning','goalkeeping_reflexes','goalkeeping_speed'
]

# Convert "xx+yy" format to numeric and fill missing values
for col in skill_cols:
    if col in df.columns:
        # Convert to numeric base value
        df[col] = df[col].astype(str).str.split('+').str[0].astype(float)
        # Fill missing values
        if col == 'goalkeeping_speed':
            # Most players are not goalkeepers → fill missing with 0
            df[col] = df[col].fillna(0)
        else:
            df[col] = df[col].fillna(df[col].median())

# Quick check
df[skill_cols].isnull().sum()
```

```python
df.set_index('long_name', inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

numeric_features = skill_cols + ['age', 'height_cm', 'weight_kg']
scaler = StandardScaler()
df_scaled = pd.DataFrame(
    scaler.fit_transform(df[numeric_features]),
    columns=numeric_features,
    index=df.index  # now using long_name
)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=10, random_state=42)
player_embeddings_pca = pd.DataFrame(
    pca.fit_transform(df_scaled),
    index=df.index,  # long_name
    columns=[f'PC{i+1}' for i in range(10)]
)
```

```python
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(n_neighbors=6, metric='euclidean')
knn.fit(player_embeddings_pca)
```

```python
def most_similar(player_name, n=5):
    if player_name not in player_embeddings_pca.index:
        return f"{player_name} not found"

    vec = player_embeddings_pca.loc[player_name].to_numpy().reshape(1, -1)
    distances, indices = knn.kneighbors(vec)

    # Skip itself
    return list(player_embeddings_pca.index[indices.flatten()[1:n+1]])
```

```python
def most_similar_partial(name, n=5):
    # Find matching names
    matches = player_embeddings_pca.index[player_embeddings_pca.index.str.contains(name, case=False)]

    if len(matches) == 0:
        return f"No players found matching '{name}'"

    # Take the first match
    player_name = matches[0]
    player_vec = player_embeddings_pca.loc[player_name].to_numpy().reshape(1, -1)
    distances, indices = knn.kneighbors(player_vec)

    similar_players = player_embeddings_pca.index[indices.flatten()[1:n+1]]  # skip itself
    return list(similar_players)
```
