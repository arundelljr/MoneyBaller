```python
df = pd.read_csv("../raw_data/FC26_20250921.csv", low_memory=False)
print(df.info())
print(df.head())
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dropna(axis=0, how='all', inplace=True)
df.dropna(axis=1, how='all', inplace=True)
```

```python
cols = ['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']
df[cols] = df[cols].fillna(0)
df.fillna({
    'overall': df['overall'].median(),
    'potential': df['potential'].median(),
}, inplace=True)
```

```python
numeric_cols = ['overall', 'potential', 'age', 'height_cm', 'weight_kg']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
print(df[numeric_cols].describe())
```

```python
import numpy as np
skill_keywords = [
    'pace','shoot','pass','drib','defend','physic',
    'attacking','skill','movement','power','mentality','goalkeeping'
]
skill_cols = [c for c in df.columns if any(k in c.lower() for k in skill_keywords)]
```

```python
df['main_position'] = df['player_positions'].str.split(',').str[0].str.strip()
```

```python
def minmax_group(x):
    return (x - x.min()) / (x.max() - x.min()) if x.max() != x.min() else 0
```

```python
df[skill_cols] = df.groupby('main_position')[skill_cols].transform(minmax_group)
```

```python
for c in skill_cols:
    df[c] = (
        df[c].astype(str)
              .replace(r'^\s*$', np.nan, regex=True)       
              .str.extract(r'(-?\d+(?:\.\d+)?)')[0]      
              .astype(float)
    )
```

```python
df['skill_score'] = df[skill_cols].mean(axis=1)
```

```python
normalized_df = df.copy()
normalized_df.head()
```

```python
physical_cols = ['height_cm', 'weight_kg', 'age', 'value_eur', 'wage_eur', 'overall', 'potential']
```

```python
for col in physical_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
for col in physical_cols:
    mean = df[col].mean(skipna=True)
    std = df[col].std(skipna=True)
    if std == 0 or np.isnan(std):  
        df[f'{col}_z'] = 0
    else:
        df[f'{col}_z'] = (df[col] - mean) / std
```

```python
df.fillna(0, inplace=True)
```

```python
df.to_csv("raw_data/normalize.csv")
```
